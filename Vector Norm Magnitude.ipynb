```python
import numpy as np
x = np.array([2,3,5])
y = np.array([3,4])
```

```python
"""For finding the norm we use NORM function from LINEAR ALGEBRA module"""
from numpy.linalg import norm

len_y = norm(y)
print(len_y)
```

```python

len_x = norm(x)
print(len_x)
```

## L1 Norm: Sum of absolute of all the elements in the vector

```python
x = np.array([-1,2,4,-2])
l1_norm = norm(x, ord = 1) # order means which norm to select. 1 =  L1 norm
l1_norm
```

## p-norm: Generalized norm
- p=1, L1 norm
- p=2, L2 norm

```python
z = np.array([1,2,-4,-5])

"""p= 1"""
p1 = norm(z,1)
print(p1)

"""p = 2"""
p2 = norm(z,2)
print(p2)

"""p = 3"""
p3 = norm(z,3)
print(p3)
```
